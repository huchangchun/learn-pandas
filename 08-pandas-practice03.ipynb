-pivot_table根据一个或多个键举行聚合，并根据行列上的分组键将数据分配到各个举行区域中
-DataFrame.pivot_table(data,
-                   values=None,
-                   index=None,
-                    columns=None,
-                    aggfunc='mean',
-                    fill_value = None,
-                    margins=False,
-                    dropna=True,
-                    margins_name='All')
-data: DataFrame对象
-values:显示的列的名字，可以应用aggfunc中的函数
-index:索引
-columns:可选的，通过额外的方法来分割你所关心的实际值
-    然而aggfunc被应用到values上，aggfunc默认的是mean

```python
#1.查看不同性别对电影的评分
df2.groupby('Gender')['Rating'].mean()
rating_by_gender = df2.pivot_table(values='Rating',index = 'Title',columns='Gender',aggfunc='mean')
rating_by_gender.head(10)
```

```python
#2.不同性别差异最大的电影
rating_by_gender['diff'] = np.fabs(rating_by_gender.F - rating_by_gender.M)
rating_by_gender.sort_values(by = 'diff',ascending = False).head(10)
```
